# Step 2: Data Preprocessing

```
#reading the data
df=pd.read_csv('Churn_Modelling.csv')
df.head()
```

```
df.shape
```

```
#selecting independent and dependent variables
df
```

```

df.drop(['RowNumber'	,'CustomerId',	'Surname','Exited'], axis=1)
```

```
x=df.drop(['RowNumber'	,'CustomerId',	'Surname','Exited'], axis=1)
```

```
x.shape
```

```
y=df['Exited']
```

```
y.head()
```

```
#independent var
x.head()
```

```
#dependent var
y.head()
```

```
#encoding categorical data - like gender and geography
from sklearn.preprocessing import LabelEncoder
```

```
label_1= LabelEncoder()
x['Geography']=label_1.fit_transform(x['Geography'])
```

```
x.head()
```

```
label_2=LabelEncoder()
x['Gender'] = label_2.fit_transform(x['Gender'])
```

```
x.head(15)
```

```
#Avoiding Dummy variable trap
x=pd.get_dummies(x, drop_first=True, columns=['Geography'])
```

```
x.head(10)
```

```
#splitting data into train and test set
from sklearn.model_selection import train_test_split
```

```
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=0)
```

```
x_train
```

```
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test =sc.transform(x_test) #no need of fit, already done in the train obj in prev step
```

```
x_train
```

```
#all values are now not having a scale difference
#now same thing
```

# Step 3: Building the Model

```
#Initializing the ANN
#creating the object

model=tf.keras.models.Sequential()
```

```
#adding first hidden input layer
#units=6 (11+1/2=6)
#activation func = relu
#input dimension=11
model.add(tf.keras.layers.Dense(units=6,activation='relu',input_dim=11))
```

```
#adding second hidden layer, no need to specify input_dim here.
model.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```
#Output layer, units=1 cuz only 1 output,activation func=sigmoid because binary output
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```
#compiling model  
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```
model.summary()
```

# Step 4: Training the model

```
#y_train is 1 column - it is a vector. whereas x_train is a numpy array so i have to convert y_train into numpy array
model.fit(x_train,y_train.to_numpy(),batch_size=32,epochs=10) 
```

# Step 5: Model evaluation and prediction

```
test_loss, test_acc = model.evaluate(x_test,y_test)
```

```
 print('Test accuracy: {}'.format(test_acc))
```

```
y_pred= model.predict_classes(x_test)
```
